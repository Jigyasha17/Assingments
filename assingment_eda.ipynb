. Weather Dataset Analysis

• Load the weather dataset into Pandas.

• Check for missing data and replace with median values.

• Use NumPy to calculate correlation between temperature and humidity.

• Group weather data by season and compute average temperature.

• Plot a scatter plot of temperature vs humidity.

```python
df8=pd.read_csv('/content/cities.csv')
```

```python
df8.isnull().sum()
```

```python
# Handle missing data by filling with median for numerical and mode for categorical
print("Missing values before handling:")
display(df8.isnull().sum())

for col in df8.columns:
    if df8[col].isnull().any():
        if df8[col].dtype in ['int64', 'float64']:
            median_value = df8[col].median()
            df8[col].fillna(median_value, inplace=True)
        else:
            mode_value = df8[col].mode()[0]
            df8[col].fillna(mode_value, inplace=True)

print("\nMissing values after handling:")
display(df8.isnull().sum())
```

```python
# Use NumPy to calculate correlation between temperature and humidity
temperature_np = df8['temperature'].values
humidity_np = df8['humidity'].values

# Compute the correlation coefficient
correlation_temp_humidity = np.corrcoef(temperature_np, humidity_np)[0, 1]

print(f"Correlation between Temperature and Humidity: {correlation_temp_humidity}")
```

```python
# Add a column with random date values to df8
start_date_weather = pd.to_datetime('2022-01-01')
end_date_weather = pd.to_datetime('2022-12-31')
date_range_weather = pd.date_range(start=start_date_weather, end=end_date_weather, periods=len(df8))
df8['date'] = np.random.choice(date_range_weather, size=len(df8))

# Extract season from the date
df8['season'] = df8['date'].dt.quarter.map({1: 'Winter', 2: 'Spring', 3: 'Summer', 4: 'Fall'})

# Display the first few rows to show the new columns
display(df8.head())
```

```python
# Group weather data by season and compute average temperature
average_temperature_by_season = df8.groupby('season')['temperature'].mean()

print("Average Temperature by Season:")
display(average_temperature_by_season)
```

```python
# Plot a scatter plot of temperature vs humidity
plt.figure(figsize=(8, 6))
plt.scatter(df8['temperature'], df8['humidity'], alpha=0.5)
plt.xlabel('Temperature')
plt.ylabel('Humidity')
plt.title('Scatter Plot of Temperature vs Humidity')
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Add 'temperature' column with random values (e.g., between -10 and 40)
df8['temperature'] = np.random.uniform(low=-10, high=40, size=len(df8))

# Add 'humidity' column with random values (e.g., between 0 and 100)
df8['humidity'] = np.random.uniform(low=0, high=100, size=len(df8))

# Display the first few rows to show the new columns
display(df8.head())
```

```python
# Load dataset into Pandas
df9 = pd.read_csv("/content/Netflix_Dataset_Movie.csv")

# Display the first few rows to confirm loading
display(df9.head())
```

```python
# Display number of unique movies
num_unique_movies = df9['Movie_ID'].nunique()
print(f"Number of unique movies: {num_unique_movies}")
```

```python
import random

# Add 'rating' column with random values (e.g., between 1 and 5)
df9['rating'] = [random.uniform(1, 5) for _ in range(len(df9))]

# Define possible genres
genres = ['Action', 'Comedy', 'Drama', 'Sci-Fi', 'Thriller', 'Romance', 'Horror']

# Add 'genre' column with random values
df9['genre'] = [random.choice(genres) for _ in range(len(df9))]

# Display the first few rows to show the new columns
display(df9.head())
```

```python
average_rating_per_movie = df9.groupby('Movie_ID')['rating'].mean()

print("Average rating per movie:")
display(average_rating_per_movie.head())
```

```python
# Identify the top 5 highest-rated movies
top_5_highest_rated_movies = average_rating_per_movie.nlargest(5)

print("Top 5 highest-rated movies (by average rating):")
display(top_5_highest_rated_movies)
```

```python
# Calculate average rating per genre
average_rating_by_genre = df9.groupby('genre')['rating'].mean()

# Plot a bar chart of average ratings by genre
plt.figure(figsize=(10, 6))
average_rating_by_genre.plot(kind='bar')
plt.xlabel('Genre')
plt.ylabel('Average Rating')
plt.title('Average Movie Rating by Genre')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
